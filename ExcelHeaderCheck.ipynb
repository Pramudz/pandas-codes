```python
import os as os
import pandas as pd
```

```python
currentpath = os.getcwd()
currentpath
```

```python
# List of Excel file paths
excel_files = []
for files in os.listdir('LOC_WISE'):
    if files.endswith(".xlsx"):
        fullFile = os.path.join(currentpath,'LOC_WISE\\'+files)
        excel_files.append(fullFile)
```

```python
# Load Excel files into DataFrames
dataframes = []
for frames in excel_files:
    readfile = pd.read_excel(frames)
    readfile.name = frames[-11:]
    dataframes.append(readfile)
```

```python
# Extract headers from each DataFrame
# headers_list = [df.columns.tolist() for df in dataframes]
check_dict = {}
for i in range(len(dataframes)):
    check_dict[dataframes[i].name] = dataframes[i].columns
    print(dataframes[i].name)
    print(dataframes[i].columns)
```

```python
matchheader = dataframes[0].columns
for key , value in check_dict.items():
    if len(value) != len(matchheader):
        print(key+"-Loss Headers")
    elif all(value == matchheader):
        print(key+"-yes")
    else:
        print(key+"-No")
```
