```python
allItems = list(demandMatrixOrig['Item Code'])
```

```python
for eachdfLoc in list(['HPC']):
    hpc = demandMatrixOrig[['Item Code','plu_code','itm_desc','Item Category','AVG Price']].copy()
    for location in list(locationMap[eachdfLoc]):
        hpc[location] = 0
    hpc.set_index('Item Code', inplace = True)  
    for itemCode in allItems:
        onhstock = stockMatrix.at[itemCode,eachdfLoc]
        if(onhstock <=0):
            continue
        for location in locationMapOrder:
            if location == eachdfLoc:
                continue
            outlet_demand = demandMatrix.at[itemCode,location]
            balance_demand= 0
            newonhstock = 0
            if outlet_demand <=0:
                continue
            if outlet_demand > onhstock:
                allocation = onhstock
                balance_demand = outlet_demand-onhstock
                newonhstock = 0
            else:
                allocation = outlet_demand
                newonhstock = onhstock-outlet_demand
                balance_demand = 0
            hpc.at[itemCode,location] = allocation
            stockMatrix.at[itemCode,eachdfLoc] = newonhstock
            demandMatrix.at[itemCode,location] = balance_demand
        hpc['total_allocation'] = hpc[list(locationMap[eachdfLoc])].sum(axis=1)
        hpc['total_value'] = hpc['total_allocation']*hpc['AVG Price']
        hpc.to_csv("allocation\\"+ eachdfLoc +".csv")
```
