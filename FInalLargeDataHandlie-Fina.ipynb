```python
import pandas as pd
df = pd.read_csv('allBuffer.csv', encoding='latin')
df.head()
```

```python
df.dtypes()
```

```python
df2 = df.loc[df['Item_Facing'] != 'Item_Facing']
df2.tail()
```

```python
df3 = df2.drop([4142358])
df3.tail()
```

```python
df3.catgry = df3.catgry.fillna('')
df3.plu_code = df3.plu_code.fillna('')
df3.Avg_7Day_sales = df3.Avg_7Day_sales.fillna(0)
df3.Item_Facing = df3.Item_Facing.fillna(0)
df3.Gondola_Stock = df3.Gondola_Stock.fillna(0)
df3.On_Hand_Quantity = df3.On_Hand_Quantity.fillna(0)
df3.UNITISE_PACK = df3.UNITISE_PACK.fillna(0)
df3.tail()
```

```python
df3['Avg_7Day_sales'] = df3['Avg_7Day_sales'].astype(float)
df3['Item_Facing'] = df3['Item_Facing'].astype(float)
df3['Gondola_Stock'] = df3['Gondola_Stock'].astype(float)
df3['UNITISE_PACK'] = df3['UNITISE_PACK'].astype(float)
df3['On_Hand_Quantity'] = df3['On_Hand_Quantity'].astype(float)
df3.tail()
```

```python
df3['total_buffer'] = df3['Item_Facing']+ df3['Gondola_Stock']
df3.tail()
```

```python
fdf = df3.loc[(df3['total_buffer'] > 0) | (df3['Avg_7Day_sales'] > 0)]
fdf.tail()
```

```python
print(fdf)
```

```python
fdf.to_csv('fillteredBuffer31stJan.csv', index=False)
```
