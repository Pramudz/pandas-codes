```python
import pandas as pd
import os
cwd = os.getcwd()
locationMap = pd.read_excel("Loc_Mapping.xlsx",index_col=None)
demandMatrixOrig = pd.read_excel("Demand and Stock.xlsx",sheet_name="Demand")
stockMatrixOrig = pd.read_excel("Demand and Stock.xlsx",sheet_name="60AgedStock")
demandMatrix = pd.read_excel("Demand and Stock.xlsx",sheet_name="Demand")
stockMatrix = pd.read_excel("Demand and Stock.xlsx",sheet_name="60AgedStock")
locationMap.head()
```

```python
locationMapOrder = locationMap.columns
```

```python
demandMatrix.head()
```

```python
stockMatrix.head()
```

```python
allItems = list(demandMatrixOrig['Item Code'])
```

```python
stockMatrix.set_index('Item Code',inplace=True)
```

```python
demandMatrix.set_index('Item Code',inplace=True)
```

```python
demandMatrix.at['AB41011B','NAWSC']
```

```python
stockMatrix.at['AB41011B','HPC']
```

```python
for location in locationMapOrder:
    eachdfLoc = location
    hpc = demandMatrixOrig[['Item Code','plu_code','itm_desc','Item Category','AVG Price']].copy()
    locationOrderBySet= list(locationMap[eachdfLoc])
    for location in locationOrderBySet:
        hpc[location] = 0
    hpc.set_index('Item Code', inplace = True)
    for itemCode in allItems:
        onhstock = stockMatrix.at[itemCode,eachdfLoc]
        if(onhstock <=0):
            continue
        for location in locationOrderBySet:
            if location == eachdfLoc:
                continue
            outlet_demand = demandMatrix.at[itemCode,location]
            balance_demand= 0
            if outlet_demand <=0:
                continue
            if outlet_demand > onhstock:
                allocation = onhstock
                balance_demand = outlet_demand-onhstock
                onhstock = 0
            else:
                allocation = outlet_demand
                onhstock = onhstock-outlet_demand
                balance_demand = 0
            hpc.at[itemCode,location] = allocation
            stockMatrix.at[itemCode,eachdfLoc] = onhstock
            demandMatrix.at[itemCode,location] = balance_demand
    hpc['total_allocation'] = hpc[list(locationMap[eachdfLoc])].sum(axis=1)
    hpc['total_value'] = hpc['total_allocation']*hpc['AVG Price']
    hpc = hpc[hpc['total_allocation'] > 0]
    hpc.to_csv("allocation\\share\\"+ eachdfLoc +".csv")
```
